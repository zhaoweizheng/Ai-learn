```python
#   np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)(, , )
x = np.linspace(-3, 3,50)
y1 = 2*x + 1
y2 = x**2

# 
plt.figure()
plt.plot(x, y1)

# 
plt.figure(num = 21, figsize=(8,5))#num , figsize 
plt.plot(x, y2)
# 
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--') 
plt.show()
```

```python
x1 = np.linspace(-3,3,50)
yx1 = 2*x1 + 1
yx2 = x1**2
plt.figure()
plt.plot(x1,yx2) # 
plt.plot(x1, yx1, color='red')
plt.xlim((-1,2)) # x, y  (-1, 2) 
plt.ylim((-2,3))
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,-1,1.22,3],
          ['really bad', 'bad', 'normal', 'good', 'very good'])

ax = plt.gca() # 
ax.spines['right'].set_color('none') // none
ax.spines['top'].set_color('none')

plt.show()
```
